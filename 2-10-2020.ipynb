# bar plot

      -->  A bar plot (or) bar graph is a chart (or) graph that presents categorical data
                with rectangular boxes 
    

```python
marks=[10,20,44,68,55]
subjects=["c","c++","java","python","oracle"]
plt.bar(subjects,marks,color=['red','green','blue','black','pink'])
```

```python
marks=[10,20,44,68,55]
subjects=["c","c++","java","python","oracle"]
plt.bar(subjects,marks,color=['red','green','blue','black','pink'],width=0.5)
```

```python
# Area plot

-> Filling the area between lines

--> legend()
    * legend creates a legend with descriptive labels for each plotted data series
          (or)
    * legend is used to specify the location of the legend...
    
```

```python
days=[1,2,3,4]
working=[8,11,9,10]
sleeping=[8,10,11,12]
plt.stackplot(days,working,sleeping,colors=['r','g'],labels=['work','rest'])
plt.legend()
plt.show()
```

```python
# pie chart
     
    *  A pie() function that generates a pie diagram representing data in an array..
    * Pie chart are best to use when you are trying to compare parts of a whole...
    
```

```python
# plotting two or more graphs on same plot
```

```python
import matplotlib.pyplot as plt
```

```python
# line 1 points
x1=[1,2,3]
y1=[2,4,1]
plt.plot(x1,y1,label="line1")   # plotting the line 1 points

# line2 points
x2=[6,7,8]
y2=[8,3,9]

plt.plot(x2,y2,label="line2")   # plotting the line 2 points

plt.xlabel('x-axis')   # naming the x-axis
plt.ylabel('y-axis')   # naming the y-axis

plt.title("Two lines in same graph") # giving a title to graph

plt.legend()  # show a legend on the plot  

plt.show()  # function to show the plot


```

```python
# scatter plot
x=[5,6,7,2,3,4,8,9,1]   # x axis values

y=[4,3,7,8,2,1,5,6,9]  # y axis values

plt.scatter(x,y,label="stars",color="green",marker='*')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("MY SCATTER PLOT")
plt.legend()

plt.show()
```

```python
# customization of plots

x=[1,2,3,4,5,6]
y=[2,4,1,5,2,6]
plt.plot(x,y,color='red',linestyle='dashed',linewidth=5,marker='o',
         markerfacecolor='green',markersize=16)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("some cool customization")
plt.show()
```

```python
# customization of plots

x=[1,2,3,4,5,6]
y=[2,4,1,5,2,6]
plt.plot(x,y,color='red',linestyle='solid',linewidth=5,marker='o',
         markerfacecolor='green',markersize=16)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("some cool customization")
plt.show()
```
